```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
# Set the style for professional appearance
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("m" \
"uted")


# my version
df_benchmark = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\extra_benchmark.csv")
df_exploration = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\extra_exploration.csv")
df_optimization = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\extra_optimization.csv")
df_narrow = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\extra_narrow.csv")
df_wide = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\extra_wide.csv")


df_benchmark["type"] = "benchmark"
df_exploration["type"] = "exploration"
df_optimization["type"] = "optimization"
df_all = pd.concat([
    df_benchmark,
    df_exploration,
    df_optimization
], ignore_index=True)
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 

# Enable LaTeX rendering
plt.rcParams['text.usetex'] = True
sns.set_theme(style="whitegrid", context="paper", font="Times New Roman")

# Create figure with shared x-axis 
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(7, 6.3), sharex=True) 

# Plot 1 (top): Cost vs Iteration 
sns.lineplot(data=df_all, x="iteration", y="cost", ax=ax1, 
             linewidth=1.3, hue="type") 
ax1.set_ylabel(r'\textbf{Cost}', fontsize=12, fontweight='semibold', labelpad=5) 
ax1.grid(True, alpha=0.3) 
ax1.spines['top'].set_visible(False) 
ax1.spines['right'].set_visible(False) 
ax1.set_xlim(0, 1000) 
ax1.set_ylim(-1, 39.0) 

# Plot 2 (bottom): Frequency vs Iteration 
sns.lineplot(data=df_all, x="iteration", y="freq", ax=ax2, 
             linewidth=1, hue="type") 
ax2.set_xlabel(r'\textbf{Iteration}', fontsize=12) 
ax2.set_ylabel(r'\textbf{Frequency}', fontsize=12, labelpad=5) 
ax2.grid(True, alpha=0.3) 
ax2.spines['top'].set_visible(False) 
ax2.spines['right'].set_visible(False) 
ax2.set_xlim(-10, 1000) 
ax2.set_ylim(-0.03, 1.0) 

# Remove the second legend
ax2.get_legend().remove() 

# Get legend handles and labels from the first axes
handles, labels = ax1.get_legend_handles_labels()

# Create LaTeX formatted labels
new_labels = [r"$\beta_0=0.1, \beta_1=5$", 
              r"$\beta_0=0.01, \beta_1=0.5$", 
              r"$\beta_0=9, \beta_1=10$"]

legend = ax1.legend(
    handles=handles,
    labels=new_labels,
    loc='upper right',
    bbox_to_anchor=(0.98, 0.98),  # Inside the plot, top-right
    frameon=True,
    fontsize=9,                   # Slightly smaller
    title=r"\textbf{Model}",
    title_fontsize=10
)

# Tidy layout 
plt.tight_layout(pad=2.0) 
fig.patch.set_facecolor('white') 
plt.savefig("plots/betas_comparison.png", bbox_inches="tight") 
plt.show()
```

#### Comparison


```python
# import comparison dataset
df_comparison_sa = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\comparison_size_data.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set academic style
sns.set_theme(style="whitegrid", context="paper", font="serif")
plt.rcParams['text.usetex'] = True

# Define specific muted colors for the two "type" categories
custom_palette = {
    "interacting": "#4878CF",  # muted blue
    "non_interacting": "#D65F5F"   # muted red
}

# Create the point plot
plt.figure(figsize=(5.6, 3.6))
sns.pointplot(
    data=df_comparison_sa,
    x="size",
    y="iterations",
    hue="type",
    palette=custom_palette,
    markers="o",
    linestyles="-",
    dodge=0.1,
    errorbar="sd",
    capsize=0.1,      
    errwidth=1)

# Customize axis and legend
plt.xlabel(r"\textbf{size}", fontsize=12)
plt.ylabel(r"\textbf{iterations}", fontsize=12)
handles, labels = plt.gca().get_legend_handles_labels()
new_labels = ["Interacting", "Non-Interacting"]
plt.legend(
    handles=handles,
    labels=new_labels,
    title="Model",
    title_fontsize=9,
    fontsize=8,
    loc="upper left",
    frameon=True,
    bbox_to_anchor=(0.02, 0.98),
    borderaxespad=0.0
)


plt.tight_layout()
plt.savefig("plots/sa_comparison.png", bbox_inches='tight')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
# Set the style for professional appearance
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("m" \
"uted")


# my version
df_benchmark = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rgd\\extra_benchmark.csv")
df_high = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rgd\\extra_high.csv")
df_low = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rgd\\extra_low.csv")

df_benchmark["type"] = "benchmark"
df_high["type"] = "high"
df_low["type"] = "low"
df_rgd_all = pd.concat([
    df_benchmark,
    df_high,
    df_low
], ignore_index=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a single figure and axis
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot 1 (Error Rate vs Epoch)
sns.lineplot(data=df_rgd_all, x="epoch", y="error_rate", ax=ax1,
             linewidth=2, hue="type")
ax1.set_ylabel('Error Rate', fontsize=12, fontweight='semibold', labelpad=15)
ax1.set_xlabel('Iteration', fontsize=12, fontweight='semibold', labelpad=15)
ax1.grid(True, alpha=0.3)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.set_xlim(0.0, 1000)
ax1.set_ylim(-0.1, 100)

# Custom legend settings for academic style
handles, labels = ax1.get_legend_handles_labels()
legend = ax1.legend(handles, labels, title="Type",
                    loc='upper left', bbox_to_anchor=(1.02, 1),
                    frameon=False, fontsize=11, title_fontsize=12)

# Optional: Set serif font (global)
plt.rcParams['font.family'] = 'serif'

# Tidy layout
plt.tight_layout(pad=2.0)
fig.patch.set_facecolor('white')
plt.show()
```
